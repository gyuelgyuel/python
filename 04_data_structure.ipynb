# 1. 문자열 메소드

### .capitalize()

문자열의 맨앞 문자 대문자로 변환

### .title()

단어별 맨앞 문자 대문자로 변환

### .upper()/lower()

문자열 모든 문자 대/소문자로 변환

### print(x)

`x`문자열 출력

### .replace(old, new[, count])

문자열 중 `old`문자열을 `new`문자열로 앞에서부터 `count` 개수만큼 교체 (`count` 미입력시 모두 교체)

### .strip([chars])

`chars`에 해당하는 모든 문자를 제거하여 반환 (`chars` 미입력시 띄어쓰기 제거) (기존 문자열 변하지 않음)
- `lstrip` : 반복되는 왼쪽만 제거, `rstrip` : 반복되는 오른쪽만 제거

### .find(x)

`x`가 문자열에 있는지 찾아 1반환, 못찾을 경우 -1 반환

### .index(x)

`x`가 처음 나오는 `index` 반환 (못찾을 경우 `error`)

### .split(x)

`x` 문자를 찾아 문자열을 두개로 나누어 문자열 배열로 반환

### .count(x)

`x` 문자(단어) 개수를 반환

# 2. 리스트 메소드

### .append(x)

`x` 값을 마지막 인자 다음에 삽입

### .extend(iterable_x)

기존 리스트에 `iterable_x`를 연결

### .insert(idx, x)

`idx` 위치에 `x`값 삽입

### .remove(x)

`x`랑 같은 값을 가지는 인자 제거

### .pop([idx=-1])

`idx`를 입력하지 않을 경우 마지막 인자 제거

### .sort([reverse=False])

기본적으로 오름차순으로 정렬
```python
list1 = list1[::-1]
```
위의 코드로 정렬은 안되지만 원래 배열을 뒤집을 수 있음

## 2.1 list copy

### copy(list)

`copy.deepcopy(list)`로 list 안에 있는 list까지 복사 가능

## 2.2 list comprehension

`list` 안에 for문을 써서 반복문 단축

# 3. 딕셔너리 메소드

### .pop(key[, default])

`key`에 해당하는 key-value를 삭제, `key`가 없을 경우 `default`를 출력(없을 경우 error출력)

### .update(key=value)

`key`에 해당하는 값을 `value`로 대체

### .get(key[, default])

`key`에 해당하는 `value`값 가져옴, `key`가 없을 경우 `default`를 출력(없을 경우 error출력)

## 3.1 dict comprehension

`dict` 안에 for문을 써서 반복문 단축
`dict2 = [e: e ** 2 for e in range(1,4))

# 4. 세트 메소드

### .add(x)

`x`를 세트에 추가

### .update(set)

`set`의 요소를 세트에 추가

### .remove(x)

`x`를 세트에서 제거, `x`가 없을 경우 error 출력

### .pop()

세트에 요소 제거 (순서 : fixed hash value에 따라 순서를 정하여 제거)

# 5. map(), zip(), filter()

- 함수를 시행하고 `list(result)`를 해야 `object`가 변경됨
- `map(function, iterable)` : `iterable` 요소들을 `function`으로 감싸 표현 (계산은 실제 호출 되었을 때 시행)
- `zip(iterable1, iterable2)` : `iterable1`과 `iterable2`의 i번째 요소를 쌍으로 묶어 i번째 요소 생성
- `filter(function, iterable)` : `True`,`False`로 반환되는 함수를 적용시켜 `True` 값을 가지는 요소들만 남김
