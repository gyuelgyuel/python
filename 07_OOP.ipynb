# 객체지향 프로그래밍(OOP)

- class : 같은 종류의 집단에 속하는 속성(attribute)과 행위(method)를 정의한 것
- instance : class를 실제로 메모리상에 할당한 것
- attribute : class/instance가 가지고 있는 데이터/값
- method : class/instance가 가지고 있는 함수/기능

## Class

### 선언

```python
class ClassName:
    attribute = value
    def method_name(self):
        code
```

### 인스턴스화

`ClassName()`을 통해 인스턴스화, 생성자 함수가 있을 경우 자동적으로 실행

### self

- `self` : instance 객체 자기자신
- 특별한 상황을 제외하고는 method의 첫번째 인자로 설정
- instance method를 실행할 때 자동으로 첫번째 인자에 instance를 할당
- 

### 생성자, 소멸자

```python
class MyClass:
    def __init__(self):
        code
    def __del__(self):
        code
```

- 각각 __init__, __del__로 함수 이름 설정
- 각각 해당 class type의 object가 생기고 삭제될 때, 함수 code 실행

### instance 변수, class 변수

- instance 변수 : instance 내부에서 생성한 변수(self.variable =)
- class 변수 : class 선언 블록 최상단에 위치, class 전체의 통합 variable (cls.variable = )
  
### class method, instance method, static method

- class method와 instance method는 `method`의 인자에 각각 cls, self가 들어가고 statc method는 특별히 인자가 필요하지 않음

### 상속

```python
class ChildClass(ParentClass):
    def child_method(self):
        code
```

class 선언시 parentclass 이름을 괄호안에 넣어주고, parentclass의 변수와 함수 사용가능



# 예시

````python
class Fan:
    power = False
    wind_level = 1
    timer = 0
    rotate = False
    def power_button(self):
        self.power = not self.power
    def wind_level_button(self):
        if self.wind_level < 3:
            self.wind_level += 1
        else:
            self.wind_level = 1
    def timer_button(self):
        if self.timer < 120:
            self.timer += 30
        else:
            self.timer = 0
    def rotate_button(self):
        self.rotate = not self.rotate
    def status(self):
        print('전원: '+str(self.power)+', 바람세기: '+str(self.wind_level)+', 타이머: '+str(self.timer)+'분, 회전: '+str(self.rotate))
print('dison\n')
dison = Fan()
dison.status()
dison.power_button()
dison.wind_level_button()
dison.wind_level_button()
dison.rotate_button()
dison.status()
print('\nlg\n')
lg = Fan()
lg.status()
lg.power_button()
lg.wind_level_button()
lg.status()
````

````python
class poketmon():
    name = ''
    skill = {}
    skill_cnt = 0
    hp = 0
    mp = 0
    type = ''
    poketmon_num = 0
    def __init__(self,name,hp,mp):
        self.name = name
        self.hp = hp
        self.mp = mp
        poketmon.poketmon_num += 1
    def attack(self,pkm,skill_name):
        dmg = self.skill[skill_name][1]
        if self.type == pkm.type:
            final_dmg = 2*dmg
        else:
            final_dmg = dmg
        pkm.hp -= final_dmg
        

class light_pkmon(poketmon):
    type = 'light'
    def learn_skill(self,skill_name,skill_type,skill_dmg):
        self.skill[skill_name] = ((skill_type,skill_dmg))
        self.skill_cnt += 1
    def info(self):
        print(f'{self.name}의 체력/마나: {self.hp}/{self.mp}, 속성: 빛, 스킬 개수: {self.skill_cnt}')

p1 = light_pkmon('pikachu',1000,1000)
p1.learn_skill('thunderbolt','light',350)
p1.info()

p2 = light_pkmon('mumu',400,500)
p2.info()

p1.attack(p2,'thunderbolt')
p2.info()

````
