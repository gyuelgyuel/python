# 1. 문자열 표시

- `'`,`"` 로 단어를 감싸서 문자열로 표시
- 여러줄 문자열을 작성할 땐 ``` or """로 감싸서 표시
- 문자열 안에서 `'`,`"`를 쓸경우 다른 `'`,`"`를 사용하거나 `\'`,`\"`를 사용

## print format

- print() string interpolation
  - % formatting : `print("sdfa %s sdf", % a) \n`
  - str.format() : `print("sdfa {} sdf".format(a))`
  - f-string : `print(f"sdfa {a} sdf")`

# 2. 문자열 메소드

## +

앞과 뒤의 str을 연결(concatenate)

## .capitalize()

문자열의 맨앞 문자 대문자로 변환

## .title()

단어별 맨앞 문자 대문자로 변환

## .upper()/lower()

문자열 모든 문자 대/소문자로 변환

## print(x)

`x`문자열 출력

## .replace(old, new[, count])

문자열 중 `old`문자열을 `new`문자열로 앞에서부터 `count` 개수만큼 교체 (`count` 미입력시 모두 교체)

## .strip([chars])

`chars`에 해당하는 모든 문자를 제거하여 반환 (`chars` 미입력시 띄어쓰기 제거) (기존 문자열 변하지 않음)
- `lstrip` : 반복되는 왼쪽만 제거, `rstrip` : 반복되는 오른쪽만 제거

## .find(x)

`x`가 문자열에 있는지 찾아 1반환, 못찾을 경우 -1 반환

## .index(x)

`x`가 처음 나오는 `index`를 `int`로 반환 (못찾을 경우 `error`)

## .split(sep = None, maxsplit = -1)

`sep`에 할당된 문자(기본 구분문자 ex.띄어쓰기,tab)로 문자열을 `maxsplit`에 할당된 갯수만큼 구분하여 `list`로 반환

## .count(x)

`x` 문자(단어) 개수를 반환
